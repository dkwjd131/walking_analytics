```python
import json
from collections import OrderedDict
from pprint import pprint

with open("palgongsan-1-Acc2-export.json",encoding='UTF-8') as json_file:
    json_acc2 = json.load(json_file)

print("가속도 센서에 대한 정보는 다음과 같습니다.", end="\n\n")
pprint(json_acc2)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(json_acc2).T
print(df)
```

```python
#time = pd.DataFrame(df.index, columns = ['time'])
time = pd.DataFrame(df.index, columns = ['time'])

dataframe = df.reset_index().rename(columns ={"index":"time"})

print(dataframe)
```

```python
#dataframe.to_excel('dataframe.xlsx')
dataframe_part= dataframe[['time','x_acc','y_acc','z_acc']]

print(dataframe_part)
```

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

dataframe_part= dataframe[['time','x_acc','y_acc','z_acc']]

plt.plot(dataframe_part.index, dataframe_part.x_acc, marker='s', color='r')
plt.plot(dataframe_part.index, dataframe_part.y_acc, marker='o', color='g')
plt.plot(dataframe_part.index, dataframe_part.z_acc, marker='*', color='b')

plt.title('Line Graph w/ different markers and colors', fontsize=20) 
plt.ylabel('value', fontsize=14)
plt.xlabel('time', fontsize=14)
plt.legend(fontsize=12, loc='best')
plt.show()
```
